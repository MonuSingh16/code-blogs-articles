```python
#### 1. Unpacking Dictionary Keys as argument:
# Each key in input_data becomes a named argument.
# Common in Pydantic, FastAPI, and dataclass initialization.

input_data = {"name": "Alice", "age": 30}
def greet(name, age):
    print(f"Hello {name}, age {age}")

greet(**input_data)  # Equivalent to greet(name="Alice", age=30)
```
