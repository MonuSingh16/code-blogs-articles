```python
lambda x : x
# Lambda functions are functions

# let's comapre a lambda function with a standard function
# lambda func are anyonymous. They don't have a name
# do not have return keyword. Whatever follows the colon is the expression

# You can have only single that evaluated to a value.
# This is value returned by lambda function

def do_boring_stuff(x):
    return 42 * x

lambda x : 42 * x
```

```python
print(do_boring_stuff(10))
print((lambda x : 42 * x)(10)) # enclose the expression in parentheses
```

```python
# To name a lambda function -
name_lambda_func = lambda x : 42 * x
print(name_lambda_func(10))
```

```python
print((lambda x, y: x * y)(10, 2))
print((lambda x, y, z: x * y + z)(10, 2, 5))
```

```python
scores = {
    "Stephen": 8,
    "James": 4,
    "Sarah": 9,
    "Kate": 6
}

def check_pass(item):
    return scores[item] >= 5

outcomes = map(check_pass, scores)
# each key in dict score is passed to check_pass

print(list(outcomes))

# Why are the dictionary keys passed to the function and not the key-value pairs?
### -> Iteration over dictionaries in Python uses the keys. This is also the case if you use a dictionary in a for loop,

outcomes = map(lambda item: scores[item] >= 5, scores)
print(list(outcomes))

# More Pythonic & Preferrable
outcomes = [scores[item] >=5 for item in scores]
print(outcomes)
```
