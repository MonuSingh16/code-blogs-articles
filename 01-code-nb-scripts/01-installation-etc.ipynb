#### Issue 1. pip not working or not found in python virtual env


- `python -m ensurepip --upgrade`
- `python -m pip install --upgrade pip`
- `python -m pip install pydantic`

If still get errors, Try deactivating and recreting your virtual environment, if it is done via python
 
 ```
 deactivate
python -m venv .venv
source .venv/bin/activate
python -m ensurepip --upgrade
python -m pip install --upgrade pip
python -m pip install pydantic
 ```

#### 2. UV Commands
```

mkdir <project_name>
cd myproject

uv init . <creates a project structure with pyproject.toml, README.md>
uv run main.py <It will create virtual environment .venv and uv.lock>
	
source .venv/bin/activate
```



1. `uv init` can be used to create a new project
2. Python project metadata is defined in a [`pyproject.toml`](https://packaging.python.org/en/latest/guides/writing-pyproject-toml/) file. uv requires this file to identify the root directory of a project.
3. uv also manages a persistent environment with the project and its dependencies in a `.venv` directory next to the `pyproject.toml`.
4. To run a command in the project environment, use `uv run`. Alternatively the project environment can be activated as normal for a virtual environment.
5. When `uv run` is invoked, it will create the project environment if it does not exist yet or ensure it is up-to-date if it exists. The project environment can also be explicitly created with `uv sync`.
6. The lockfile is [automatically created and updated](https://docs.astral.sh/uv/concepts/projects/sync/#automatic-lock-and-sync) during uv invocations that use the project environment, i.e., `uv sync` and `uv run`. The lockfile may also be explicitly updated using `uv lock`.
