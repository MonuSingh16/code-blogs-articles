```python
team_members = [
    ("Kate", 8),
    ("John", 5),
    ("Alice, 10"),
    ("Bob, 7"),
    ("Eve, 6"),
    ("Charlie, 9"),
]

team_members[3]

# You can get the fouthr item using operator.itemgetter()
# Returns a callable, an then you can pass the list to that callable

from operator import itemgetter
get_fourth = itemgetter(3) # Returns a callable
get_fourth(team_members) # will always fetch fourth item

# Or whatever subscriptable if you can use square bracket notations to fetch

```

```python
class Team:
    def __init__(self, team_iterable):
        self.team = list(team_iterable)
    
    def __getitem__(self, key_or_index):
        print("Hey look ! __getitem__ was called behind the scenes")
        return self.team[key_or_index]
    
team = Team(team_members)
team[3]
```

```python
team_members = [
    ("Kate", 8),
    ("John", 5),
    ("Alice", 10),
    ("Bob", 7),
    ("Eve",6)
 ]

# Key parameter that you can use to set your own sorting rules
# Argument you pass to this key parameter should be function (Callable)
sorted(team_members, key=lambda x : x[1])

from operator import itemgetter
get_score = itemgetter(1)
sorted(team_members, key=get_score)
```

```python
import timeit
timeit.timeit(
    "sorted(team_members, key=get_score)",
    number=1_000_000,
    globals=globals()
)
```

```python

timeit.timeit(
    "sorted(team_members, key=itemgetter(1))",
    number=1_000_000,
    globals=globals()
)
```

```python
timeit.timeit(
    "sorted(team_members, key=lambda x: x[1])",
    number=1_000_000,
    globals=globals()
)
```

```python
members = [
    {"name": "Josh", "membership_tier": 1, "year_joined": 2018},
     {"name": "Sally", "membership_tier": 2, "year_joined": 2020},
     {"name": "Mike", "membership_tier": 1, "year_joined": 2019},
     {"name": "Anna", "membership_tier": 2, "year_joined": 2017},
     {"name": "Tom", "membership_tier": 1, "year_joined": 2018},
     {"name": "Lucy", "membership_tier": 2, "year_joined": 2020},
 ]

from operator import itemgetter
sort_order = itemgetter("membership_tier", "year_joined", "name")
# can accept multiple arguments.

sort_order(members[0])
# returns a tuple with values in order of keys

sorted(members, key=sort_order)
```
