```python
team_members = [
    ("Kate", 8),
    ("John", 5),
    ("Alice, 10"),
    ("Bob, 7"),
    ("Eve, 6"),
    ("Charlie, 9"),
]

team_members[3]

# You can get the fouthr item using operator.itemgetter()
# Returns a callable, an then you can pass the list to that callable

from operator import itemgetter
get_fourth = itemgetter(3) # Returns a callable
get_fourth(team_members) # will always fetch fourth item

# Or whatever subscriptable if you can use square bracket notations to fetch

```

```python
class Team:
    def __init__(self, team_iterable):
        self.team = list(team_iterable)
    
    def __getitem__(self, key_or_index):
        print("Hey look ! __getitem__ was called behind the scenes")
        return self.team[key_or_index]
    
team = Team(team_members)
team[3]
```
