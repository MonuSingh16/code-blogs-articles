#### 2. UV Commands
```

mkdir <project_name>
cd myproject

uv init . <creates a project structure with pyproject.toml, README.md>
uv run main.py <It will create virtual environment .venv and uv.lock>
	
source .venv/bin/activate
```



1. `uv init` can be used to create a new project
2. Python project metadata is defined in a [`pyproject.toml`](https://packaging.python.org/en/latest/guides/writing-pyproject-toml/) file. uv requires this file to identify the root directory of a project.
3. uv also manages a persistent environment with the project and its dependencies in a `.venv` directory next to the `pyproject.toml`.
4. To run a command in the project environment, use `uv run`. Alternatively the project environment can be activated as normal for a virtual environment.
5. When `uv run` is invoked, it will create the project environment if it does not exist yet or ensure it is up-to-date if it exists. The project environment can also be explicitly created with `uv sync`.
6. The lockfile is [automatically created and updated](https://docs.astral.sh/uv/concepts/projects/sync/#automatic-lock-and-sync) during uv invocations that use the project environment, i.e., `uv sync` and `uv run`. The lockfile may also be explicitly updated using `uv lock`.

#### 3. Installation Issue - Package Dependency - lamini / u-map

##### The issue -

No solution found when resolving dependencies:

  ╰─▶ Because umap was not found in the package registry and lamini==3.4.6 depends on umap, we can conclude that lamini==3.4.6 cannot be used.
      And because only lamini<=3.4.6 is available and your project depends on lamini>=3.4.6, we can conclude that your project's requirements are unsatisfiable.

```
// ...existing code...
[tool.uv]
override-dependencies = [
    "umap ; python_version < '0'",
]
// ...existing code...
```
1. The override-dependencies entry tells `uv`: "Whenever any package asks for umap, only try to install it if the Python version is less than 0."
2. Since the Python version is never less than 0, `uv` effectively deletes the `umap` dependency from the resolution tree.
3. By adding `umap-learn`, you provide the actual library that `lamini` needs at runtime (both provide the import `umap` module).

```
# Add umap-learn manually so the code actually works
uv add umap-learn

# Install the desired package
uv add reasoning_from_scratch
```

#### 4. NLTK Compatability Issue with Python

AttributeError: partially initialized module 'nltk' has no attribute 'data' (most likely due to a circular import)

NLTK's downloader.py tries to initialize the Downloader class during module load, which references nltk.data.path before the data module is fully initialized.

Upgrade nltk to a comptabile version `uv add nltk==3.9.1`, then restart the kernel to the same update to take effect as sometime the effect does not take place.
